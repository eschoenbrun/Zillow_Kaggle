```python
import h2o
import os
```

```python
h2o.init(max_mem_size="4G")
h2o.remove_all()  
```

```python
from h2o.estimators.random_forest import H2ORandomForestEstimator
```

```python
df = h2o.import_file(os.path.realpath("data/join_after_imp.csv"))
```

```python
train, valid, test = df.split_frame([0.6, 0.2], seed=4)
```

```python
len(train), len(valid),len(test)
```

```python
#Prepare predictors and response columns
df_X = df.col_names[:-1]     #last column is Cover_Type, our desired response variable 
df_y = df.col_names[-1] 
```

```python
df_y
```

```python
rf_v1 = H2ORandomForestEstimator(
    model_id="rf_df_v1",
    ntrees=200,
    stopping_rounds=2,
    score_each_iteration=True,
    seed=1000000)
```

```python
rf_v1.train(df_X, df_y, training_frame=train, validation_frame=valid)
```

```python
rf_v1
```

```python
rf_v1.score_history()
```

## Gradient Boosting

```python
from h2o.estimators.gbm import H2OGradientBoostingEstimator
```

```python
gbm_v1 = H2OGradientBoostingEstimator(
    model_id="gbm_df_v1",
    seed=4
)
```

```python
gbm_v1.train(df_X, df_y, training_frame=train, validation_frame=valid)
```

```python
gbm_v1
```

```python
gbm_v1.score_history()
```

```python
gbm_v2 = H2OGradientBoostingEstimator(
    ntrees=200,
    learn_rate=0.2,
    max_depth=10,
    stopping_tolerance=0.01, #10-fold increase in threshold as defined in rf_v1
    stopping_rounds=2,
    score_each_iteration=True,
    model_id="gbm_df_v2",
    sample_rate=0.7,
    seed=2000000
)
```

```python
gbm_v2.train(df_X, df_y, training_frame=train, validation_frame=valid)
```

```python
gbm_v2
```

```python
final_rf_predictions = gbm_v2.predict(test[:-1])
```
